```python
import os
import pandas as pd

book_dir = "./books"

# Initialize an empty list to store data
data = []

# Loop through each file in the books directory
for book in os.listdir(book_dir):
    try:
        with open(os.path.join(book_dir, book), "r") as file:
            text = file.read()
        data.append({'Book': book, 'Text': text})
    except Exception as e:
        print("Error with book:", book, "\nError:", e)

# Create a DataFrame from the data list
books_df = pd.DataFrame(data)

books_df
```

```python
def trim_text_after_dash(text):
    return text.split("---")[0]

# Apply the function to the 'Text' column
books_df['Text'] = books_df['Text'].apply(trim_text_after_dash)

books_df.head().Text[0]
```

```python
def extract_author_title(text):
    lines = text.split('\n', 5)[:5]
    author_title = ' '.join(lines).strip()
    return author_title

# Apply the function to the 'Text' column
books_df['Book'] = books_df['Text'].apply(extract_author_title)

books_df.head()
```

```python
# save the dataframe to a csv file
books_df.to_csv('books.csv', index=False)
```
